```python
import matplotlib.pyplot as plt 
import cartopy.crs as ccrs
import cartopy.feature as cfeat

%matplotlib inline
```

```python
fig = plt.figure(figsize=(15,15))
ax  = fig.add_subplot(1, 1, 1, projection=ccrs.Mercator())
ax.add_feature(cfeat.LAND)
ax.add_feature(cfeat.OCEAN)
ax.add_feature(cfeat.COASTLINE)
ax.add_feature(cfeat.BORDERS, linestyle = ':')
ax.add_feature(cfeat.LAKES, alpha = 0.5)
ax.add_feature(cfeat.RIVERS)

#ax.set_extent([-100, -10, 5, 40])
ax.set_extent([-75, -65, 15, 22])
```

```python
import cartopy.io.shapereader as shpreader
import cartopy.feature as cfeature

fig = plt.figure(figsize=(10, 6))
fig = plt.axes(projection=ccrs.PlateCarree())
fig.set_extent([-75, -65, 15, 22])

DIR = '../shape_files'
Features = ['ne_10m_admin_1_states_provinces/ne_10m_admin_1_states_provinces.shp',
            'ne_10m_land/ne_10m_land.shp', 'ne_10m_ocean/ne_10m_ocean.shp',
            'ne_10m_coastline/ne_10m_coastline.shp']

for Feature in Features:
    #print(f'reading: {Features}')
    reader = shpreader.Reader(f'{DIR}/{Feature}')
    counties = list(reader.geometries())
    COUNTIES = cfeature.ShapelyFeature(counties, ccrs.PlateCarree())
    fig.add_feature(COUNTIES, facecolor='none', edgecolor='gray')


fig
```

```python
import cartopy.io.shapereader as shpreader
import cartopy.feature as cfeature
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER

plt.figure(figsize=(10, 6))
fig2 = plt.axes(projection=ccrs.PlateCarree())
fig2.set_extent([-72, -68, 17.5, 20])

DIR = '../shape_files'
Features = ['ne_10m_admin_1_states_provinces/ne_10m_admin_1_states_provinces.shp',
            'ne_10m_land/ne_10m_land.shp', 'ne_10m_ocean/ne_10m_ocean.shp',
            'ne_10m_coastline/ne_10m_coastline.shp']

reader = shpreader.Reader(f'{DIR}/{Features[0]}')
counties = list(reader.geometries())
COUNTIES = cfeature.ShapelyFeature(counties, ccrs.PlateCarree())
fig2.add_feature(COUNTIES, facecolor='none', edgecolor='gray')

reader = shpreader.Reader(f'{DIR}/{Features[1]}')
counties = list(reader.geometries())
COUNTIES = cfeature.ShapelyFeature(counties, ccrs.PlateCarree())
fig2.add_feature(COUNTIES, facecolor='gray', edgecolor='gray', zorder = -1, alpha = 0.1)

reader = shpreader.Reader(f'{DIR}/{Features[3]}')
counties = list(reader.geometries())
COUNTIES = cfeature.ShapelyFeature(counties, ccrs.PlateCarree())
fig2.add_feature(COUNTIES, facecolor='gray', edgecolor='gray', zorder = -1, alpha = 0.1)



fig2.add_feature(cfeature.OCEAN)
# grid_lines = fig2.gridlines(draw_labels=True)
# grid_lines.xformatter = LONGITUDE_FORMATTER
# grid_lines.yformatter = LATITUDE_FORMATTER

fig2.set_extent([-72, -70, 18, 20])
```
